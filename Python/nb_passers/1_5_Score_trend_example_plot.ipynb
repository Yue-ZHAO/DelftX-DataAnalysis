```python
# Data Generation
columns = ["Week", "Unstable Scores", "Chasing Scores", "Reducing Scores", "Keeping High Scores", "Keeping Mid Scores"]
df1 = pd.DataFrame(columns=columns)
```

```python
# Keeping mid score
df1 = df1.append({'Week': 1, 'Unstable Scores': 0.9, 'Chasing Scores': 0.5, 'Reducing Scores': 1,    'Keeping High Scores': 0.93, 'Keeping Mid Scores': 0.75}, ignore_index=True)
df1 = df1.append({'Week': 2, 'Unstable Scores': 0.2, 'Chasing Scores': 0.6, 'Reducing Scores': 0.96, 'Keeping High Scores': 0.95, 'Keeping Mid Scores': 0.60}, ignore_index=True)
df1 = df1.append({'Week': 3, 'Unstable Scores': 0.95, 'Chasing Scores': 0.8, 'Reducing Scores': 0.82, 'Keeping High Scores': 0.87, 'Keeping Mid Scores': 0.68}, ignore_index=True)
df1 = df1.append({'Week': 4, 'Unstable Scores': 0.35, 'Chasing Scores': 0.9, 'Reducing Scores': 0.70, 'Keeping High Scores': 0.90, 'Keeping Mid Scores': 0.70}, ignore_index=True)
df1 = df1.append({'Week': 5, 'Unstable Scores': 0.9, 'Chasing Scores': 1,   'Reducing Scores': 0.58, 'Keeping High Scores': 0.92, 'Keeping Mid Scores': 0.77}, ignore_index=True)
df1['Week'] = df1['Week'].astype(int)
print df1
```

```python
# Font Size
SIZE = 20
plt.rc('font', size=SIZE)  # controls default text sizes
plt.rc('axes', titlesize=SIZE)  # fontsize of the axes title
plt.rc('axes', labelsize=SIZE)  # fontsize of the x any y labels
plt.rc('xtick', labelsize=SIZE)  # fontsize of the tick labels
plt.rc('ytick', labelsize=SIZE)  # fontsize of the tick labels
plt.rc('legend', fontsize=SIZE)  # legend fontsize
plt.rc('figure', titlesize=SIZE)  # # size of the figure title
```

```python
with plt.xkcd():
    # plt.plot(radius, area, label='Circle')
    # plt.plot(radius, square, marker='o', linestyle='--', color='r', label='Square')
    plt.plot(df1['Week'], df1['Unstable Scores'], color="blue", linewidth=2, linestyle="-", label='Unstable Scores')
    plt.plot(df1['Week'], df1['Chasing Scores'], color="red", linewidth=2, linestyle="-", label='Chasing Scores')
    plt.plot(df1['Week'], df1['Reducing Scores'], color="yellow", linewidth=2, linestyle="-", label='Reducing Scores')
    plt.plot(df1['Week'], df1['Keeping High Scores'], color="green", linewidth=2, linestyle="-", label='Keeping High Scores')
    plt.plot(df1['Week'], df1['Keeping Mid Scores'], color="grey", linewidth=2, linestyle="-", label='Keeping Mid Scores')

    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.ylabel('Score')
    plt.xlabel('Unit')
    plt.xlim(0.5, 5.5)
    plt.ylim(0, 1.05)
    xint = [1, 2, 3, 4, 5]
    plt.xticks(xint)
```

## Examples of Post-passing Score
Unstable scores, Chasing scores, Reducing scores, Keeping high scores, Keeping mid scores, and Keeping low scores.

```python
# DF initial
columns = ["Week", "Unstable Scores", "Chasing Scores", "Reducing Scores", "Keeping High Scores", "Keeping Mid Scores", "Keeping Low Scores"]
df2 = pd.DataFrame(columns=columns)
```

```python
# Data Generation
df2 = df2.append({'Week': 6, 'Unstable Scores': 0.2,  'Chasing Scores': 0.75,  'Reducing Scores': 0.96, 'Keeping High Scores': 0.93, 'Keeping Mid Scores': 0.46, 'Keeping Low Scores': 0}, ignore_index=True)
df2 = df2.append({'Week': 7, 'Unstable Scores': 0.95, 'Chasing Scores': 0.88, 'Reducing Scores': 0.41, 'Keeping High Scores': 0.95, 'Keeping Mid Scores': 0.53, 'Keeping Low Scores': 0.13}, ignore_index=True)
df2 = df2.append({'Week': 8, 'Unstable Scores': 0.1,  'Chasing Scores': 0.98, 'Reducing Scores': 0.12, 'Keeping High Scores': 0.87, 'Keeping Mid Scores': 0.48, 'Keeping Low Scores': 0.05}, ignore_index=True)

df2['Week'] = df2['Week'].astype(int)
print df2
```

```python
with plt.xkcd():
    # plt.plot(radius, area, label='Circle')
    # plt.plot(radius, square, marker='o', linestyle='--', color='r', label='Square')
    plt.plot(df2['Week'], df2['Unstable Scores'], color="blue", linewidth=2, linestyle="-", label='Unstable Scores')
    plt.plot(df2['Week'], df2['Chasing Scores'], color="red", linewidth=2, linestyle="-", label='Chasing Scores')
    plt.plot(df2['Week'], df2['Reducing Scores'], color="yellow", linewidth=2, linestyle="-", label='Reducing Scores')
    plt.plot(df2['Week'], df2['Keeping High Scores'], color="green", linewidth=2, linestyle="-", label='Keeping High Scores')
    plt.plot(df2['Week'], df2['Keeping Mid Scores'], color="grey", linewidth=2, linestyle="-", label='Keeping Mid Scores')
    plt.plot(df2['Week'], df2['Keeping Low Scores'], color="black", linewidth=2, linestyle="-", label='Keeping Low Scores')

    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.ylabel('Score')
    plt.xlabel('Unit')
    plt.xlim(5.5, 8.5)
    plt.ylim(0, 1.05)
    xint = [6, 7, 8]
    plt.xticks(xint)
```
